```python
import requests

api_key = "928b2cc2b6e2289d113ec443c573019a"  
city = "Mumbai"
url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

response = requests.get(url)
data = response.json()
print(data)
```

```python
temperature = data['main']['temp']
humidity = data['main']['humidity']
pressure = data['main']['pressure']
weather = data['weather'][0]['description']

print(f"City: {city}")
print(f"Temperature: {temperature}°C")
print(f"Humidity: {humidity}%")
print(f"Pressure: {pressure} hPa")
print(f"Weather: {weather}")
```

```python
cities = ["Mumbai", "Delhi", "New York", "Tokyo", "Sydney", "London", "Paris", "Cape Town", "Dubai", "Singapore"]
temps = []

for city in cities:
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    data = response.json()
    temps.append(data['main']['temp'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(cities, temps, color='skyblue')
plt.title("Temperature in Various Cities")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

plt.figure(figsize=(10,5))
df = pd.DataFrame({'City': cities, 'Temperature': temps})
sns.barplot(x='City', y='Temperature', data=df)
plt.title("City-wise Temperature")
plt.show()
```
